```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification", device=0)
```

```python
headlines=pd.read_csv('../input/million-headlines/abcnews-date-text.csv')
```

```python
headlines.head()
```

```python
samples=headlines['headline_text'][:100]
samples
```

```python
strings = samples.values
```

```python
sequence = list(strings)
```

# Classification

```python
candidate_labels = ["politics", "finance", "sports",'entertainment']
```

```python
classifier(sequence, candidate_labels)
```

# Sentiment Analysis

```python
sentiment_labels = ['positive','negative']
```

```python
sent = classifier(sequence, sentiment_labels)
```
